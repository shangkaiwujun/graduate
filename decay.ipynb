```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def c_0(t, lambda_, gamma):
    alpha = np.sqrt(lambda_/gamma)
    exp_part = np.exp(-1/2*alpha**2*gamma*t)
    sin_part = np.cos(1/np.sqrt(2)*alpha*gamma*t)+1/np.sqrt(2)*alpha*np.sin(1/np.sqrt(2)*alpha*gamma*t)
    return exp_part*sin_part
```

```python
lambda_ = 0.015
gamma = 0.1

# 生成 x 值的范围，例如 -10 到 10，间隔为 0.1
x_values = np.arange(0, 200, 0.1)

# 计算对应 x 值的 y 值
y_values = c_0(x_values, lambda_=lambda_, gamma=gamma)**2


# 绘制图表
plt.plot(x_values, y_values*0, c='black')
plt.plot(x_values, y_values, label=r'$\lambda$ = {0:,.3f}, $\gamma$ = {1:,.3f}'.format(lambda_, gamma), c='black')
ax = plt.gca()

# 隐藏右边和上边的边框
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#plt.title('$\|C_0\|^2$')
plt.xlabel(r'$\tau$/day')
plt.ylabel(r'$\|C_0\|^2$')
#plt.legend()  # 添加图例
#plt.grid(True)  # 添加网格
plt.show()
```
