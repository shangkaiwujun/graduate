```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def c_0(t, alpha):
    exp_part = np.exp(-1/2*alpha**2*t)
    sin_part = np.cos(1/np.sqrt(2)*alpha*t)+1/np.sqrt(2)*alpha*np.sin(1/np.sqrt(2)*alpha*t)
    return exp_part*sin_part
```

```python
def c_0_real(t, alpha):
    exp_part = np.exp(-1/2*alpha**2*t)
    D = np.sqrt(alpha**2*2-alpha**4)
    sin_part = np.cos(1/2*D*t)+1/D*alpha**2*np.sin(1/2*D*t)
    return exp_part*sin_part
```

```python
def c_0_flat(t, alpha):
    exp_part = np.exp(-1/2*alpha**2*t)
    sin_part = np.cos(1/np.sqrt(2)*alpha*t)+1/np.sqrt(2)*alpha*np.sin(1/np.sqrt(2)*alpha*t)
    return exp_part
```

```python
def concurrence_c(t, c_0_func, alpha_00, alpha):
    beta_11 = np.sqrt(1 - alpha_00**2)
    c = c_0_func(t, alpha)
    kai_2 = (1-c**2)
    lambda_ = beta_11**2*c**2*kai_2 - np.abs(alpha_00*beta_11)*c**2
    res = -2*lambda_
    return res*(res>0)
```

```python
def concurrence_r(t, c_0_func, alpha_00, alpha):
    beta_11 = np.sqrt(1 - alpha_00**2)
    c = c_0_func(t, alpha)
    kai_2 = (1-c**2)
    lambda_ = beta_11**2*c**2*kai_2 - np.abs(alpha_00*beta_11)*kai_2
    res = -2*lambda_
    return res*(res>0)
```

```python
def concurrence_c1_r1(t, c_0_func, alpha_00, alpha):
    beta_11 = np.sqrt(1 - alpha_00**2)
    c = c_0_func(t, alpha)
    kai_2 = (1-c**2)
    lambda_ = -2*beta_11**2*np.sqrt(c**2*kai_2)
    res = -2*lambda_
    return res*(res>0)
```

```python
alpha_00 = np.sqrt(1)/np.sqrt(6)
alpha = 1/np.sqrt(400)


# 生成 x 值的范围，例如 -10 到 10，间隔为 0.1
x_values = np.arange(0, 3/alpha**2, 0.1)

# 计算对应 x 值的 y 值
y_values_1 = concurrence_c(x_values, c_0_func=c_0, alpha_00=alpha_00, alpha=alpha)
y_values_2 = concurrence_r(x_values, c_0_func=c_0, alpha_00=alpha_00, alpha=alpha)
y_values_3 = concurrence_c1_r1(x_values, c_0_func=c_0, alpha_00=alpha_00, alpha=alpha)


# 绘制图表
plt.plot(x_values*alpha**2, y_values_1, label='$concurrence_{c_1c_2}$', c='black')
#plt.plot(x_values*alpha**2, y_values_2, label='$concurrence_{r_1r_2}$')
ax = plt.gca()

# 隐藏右边和上边的边框
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#plt.title('Concurrence')
plt.xlabel('$\kappa$t')
plt.ylabel('concurrence')
#plt.grid(True)  # 添加网格
#plt.legend()  # 添加图例
plt.show()

#plt.plot(x_values*alpha**2, y_values_1, label='$concurrence_{c_1c_2}$')
plt.plot(x_values*alpha**2, y_values_2, label='$concurrence_{b_1b_2}$', c='black')
ax = plt.gca()

# 隐藏右边和上边的边框
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#plt.title('Concurrence')
plt.xlabel('$\kappa$t')
plt.ylabel('concurrence')
#plt.grid(True)  # 添加网格
#plt.legend()  # 添加图例
plt.show()

plt.plot(x_values*alpha**2, y_values_3, label='$concurrence_{c_1b_1}$', c='black')
ax = plt.gca()

# 隐藏右边和上边的边框
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#plt.title('Concurrence')
plt.xlabel('$\kappa$t')
plt.ylabel('concurrence')
#plt.grid(True)  # 添加网格
#plt.legend()  # 添加图例
plt.show()
```
